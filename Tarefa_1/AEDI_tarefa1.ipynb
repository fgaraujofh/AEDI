### Elaboração do problema
Um gerente de produção em uma empresa que fabrica peças eletrônicas tem que decidir quantas peças eletrônicas produzir no próximo trimestre.
A demanda por essas peças pode variar bastante, mas o custo fixo de produção deve ser **10.000,00** reais por trimestre e o custo variável por unidade produzida é de **100,00** reais.
A estimativa média de demandas é de **500** peças e o desvio padrão é de **100** unidades.

Qual a chance de se ober um lucro positivo, ser o valor final por peça for **130,00** reais?
### Resolução do problema

```python

# Importanto bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Variáveis do Problema
custo_fixo = 10000
custo_variavel = 100
valor_final = 130
media_demanda = 500
desvio_padrao_demanda = 100
tamanho_amostra = 10000
```

```python
# Gerar amostra aleatórios da demanda estimada usando distribuição Normal
np.random.seed(27)
demandas = np.random.normal(media_demanda, desvio_padrao_demanda, tamanho_amostra)
demandas
```

```python
#  calcular o lucro para cada valor de demanda gerado
lucros = (demandas * (valor_final - custo_variavel)) - custo_fixo
lucros
```

```python
# calcular a probabilidade de obter um lucro positivo
probabilidade_lucro_positivo = np.mean(lucros > 0)
print(f"A probabilidade de obter um lucro positivo é de {probabilidade_lucro_positivo:.2%}")
```

### Descrevendo a solução
Para a resolução do problema, a **simulação Monte Carlo** foi usada para gerar valores aleatórios (amastra) da demanda de peças eletrônicas, usando-se como parâmetros a estimativa de média $\mu$ e do desvio padrão $\sigma$, informados no enunciado.

Logo após calculou-se o lucro correspondente a dada demanda estimada. Para todos os lucros calculados, foi calculada a probabilidade de se obter um lucro positivo.

```python
# Distribuição dos valores gerados por demanda
plt.hist(demandas, bins=50, density=True)
plt.title("Distribuição de demanda")
plt.xlabel("Demandas")
plt.ylabel("Densidade")

# Função de densidade de probabilidade da distribuição normal
x = np.linspace(media_demanda - 3 * desvio_padrao_demanda, media_demanda + 3 * desvio_padrao_demanda, 100)
pdf = 1 / (desvio_padrao_demanda * np.sqrt(2 * np.pi)) * np.exp(-(x - media_demanda) ** 2 / (2 * desvio_padrao_demanda ** 2))
plt.plot(x, pdf, color='red')

plt.show()
```

A opção em se escolher a distribuição normal para a geração das amostras se deu pelo fato de se ter no enunciado os parâmetros da estimativa de média $\mu$ e do desvio padrão $\sigma$

Ademais, percebe-se que a escolha da distribuição normal foi adequada para a geração de valores aleatórios para a demanda de peças eletrônicas, já que **a forma do histograma dos valores gerados aleatoriamente**, comparada com a **curva de densidade de probabilidade da distribuição normal**, usando-se a mesma média e desvio padrão, **assemelham-se muito**.
